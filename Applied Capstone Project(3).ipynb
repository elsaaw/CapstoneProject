```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
print("Hello Capstone Project")
```

```python
corona_dataset_csv = pd.read_csv('https://raw.githubusercontent.com/elsaaw/CapstoneProject/master/covid19_Confirmed_dataset.csv')
corona_dataset_csv.head(20)
```

```python
corona_dataset_csv.shape
```

```python
corona_case_agg = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_case_agg.shape
```

```python
corona_case_agg.rename({"4/30/20":"Total cases"}, axis=1, inplace=True)
```

```python
corona_splice = pd.DataFrame(corona_case_agg['Total cases'])
corona_splice.head(20)
```

```python
corona_death = pd.read_csv('https://raw.githubusercontent.com/elsaaw/CapstoneProject/master/covid19_deaths_dataset.csv')
corona_death.head()
```

```python
corona_death_agg = corona_death.groupby("Country/Region").sum()
```

```python
corona_death_agg.rename({'4/30/20': "Total deaths"}, axis=1, inplace = True)
corona_death_agg.head()
```

```python
corona_death_splice = pd.DataFrame(corona_death_agg["Total death"])
corona_death_splice.head()
```

```python
happiness_csv = pd.read_csv('https://raw.githubusercontent.com/elsaaw/CapstoneProject/master/worldwide_happiness_report.csv')
happiness_csv.head()
```

```python
happiness_csv.drop(['Overall rank'], axis=1, inplace=True)
```

```python
happiness_csv.rename({'Country or region' : 'Country/Region'}, axis = 1, inplace=True)
happiness_csv.head()
```

```python
happiness_csv.set_index(["Country/Region"], inplace=True)
```

```python
happiness_csv.head()
```
